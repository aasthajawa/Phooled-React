```python
import numpy as np
import matplotlib.pyplot as plt

# Define the categories and scores
categories = ['User-Interface', 'User-Friendliness', 'Lines of Code', 'Code-Efficiency', 
              'Structure of Code', 'Code Maintainability', 'Database Implementation', 
              'Frontend-Backend Communication', 'Routing', 'State Management', 
              'Multiplayer Implementation', 'Testing','Server Deployment',  'Development Pipeline']

react_scores = [8.375, 8.5, 7, 8.25, 9, 8, 5.7, 8.5, 8, 8.8, 8.2, 8.7, 8, 8.5]
django_scores = [5.25, 6.34, 6, 7.9, 5.5, 5.12, 9.2, 7.4, 8.2, 6.8, 7, 7.76, 7.7, 7.9]

# Calculate the number of categories
num_categories = len(categories)

# Calculate the angle for each category
angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()

# Close the plot by repeating the first data point
react_scores += react_scores[:1]
django_scores += django_scores[:1]
angles += angles[:1]

# Plot the radar chart
fig, ax = plt.subplots(figsize=(5.4, 5.4), subplot_kw=dict(polar=True))
ax.fill(angles, react_scores, color='blue', alpha=0.25)
ax.fill(angles, django_scores, color='red', alpha=0.25)

# Add labels and legend
plt.xticks(angles[:-1], categories, color='black', size=8)
plt.title('Comparison of React and Django Frameworks', size=20, color='black', y=1.1)
plt.legend(['React', 'Django'], loc='upper right', bbox_to_anchor=(1.4, 1.05))


plt.savefig('radar1')
```
